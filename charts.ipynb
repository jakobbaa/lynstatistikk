```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Data
companies = [
    "Saudi Aramco",
    "Gazprom",
    "ExxonMobil",
    "Abu Dhabi NOC",
    "TotalEnergies"
]
emissions = [441756, 283407, 246883, 131020, 117800]
reporting_status = [False, False, True, True, True]

# Farge og størrelse
colors = ["#e63946" if not reported else "#2a9d8f" for reported in reporting_status]
sizes = [e / 200 for e in emissions]

# Figur
fig, ax = plt.subplots(figsize=(7, 2.8))  # Kompakt figur

# Plot bobler og tekst
for i, (company, size, color) in enumerate(zip(companies, sizes, colors)):
    ax.plot(i, 0, marker='o', markersize=size**0.5, color=color, alpha=0.8)
    ax.text(i, -0.06, company, ha='center', va='top', fontsize=8)
    ax.text(i, -0.12, f"{emissions[i]:,} tonn", ha='center', va='top', fontsize=7)

# Aksestil
ax.set_xlim(-0.5, len(companies) - 0.5)
ax.set_ylim(-0.25, 0.15)
ax.axis('off')

# Tegnforklaring (nærmere)
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Ingen rapportering',
           markerfacecolor='#e63946', markersize=8),
    Line2D([0], [0], marker='o', color='w', label='Rapporterer til OGMP 2.0',
           markerfacecolor='#2a9d8f', markersize=8)
]
ax.legend(handles=legend_elements, loc='lower center', bbox_to_anchor=(0.5, -0.05), ncol=2, fontsize=8)

plt.tight_layout()
plt.show()
```
